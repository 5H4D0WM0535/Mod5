## Challenge 4: Count Elements in List

Write a function `count_elements(lst, x)` that returns the count of the element `x` in the list `lst`.

### Input
<code>

</code>

### Output
<code>

</code>

## Challenge 5: Remove Duplicates from List

### Input

<code>

</code>

## Challenge 7: Check Substring

## Challenge 8: Find Intersection of Two Lists

Write a function `intersection(lst1, lst2)` that returns a list of elements that are common in both `lst1` and `lst2`.

### Input
<code>

</code>

</code>
